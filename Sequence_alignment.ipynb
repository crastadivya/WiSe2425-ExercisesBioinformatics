```python
# Needleman Wunsch Algorithm

def needleman_wunsch(w1, w2, gap):
    #if len(w1)<len(w2):
     #   w1, w2 = w2, w1
    
    print("Global Alignment using Needleman-Wunsch Algorithm")
    print(f"words: {w1}, {w2} and gap: {gap}")

    score = np.zeros((len(w2) + 1, len(w1) + 1))
    arrow = np.array([['' for _ in range(len(w1)+1)] for _ in range(len(w2)+1)], dtype=object)
    arrow[0, :] = np.array(['left'] * (len(w1)+1))
    arrow[:,0] = np.array(['top'] * (len(w2)+1))
    arrow[0,0] = ''
    arrow2 = arrow.copy()
    score[0,:] = [0 + i*gap for i in range(len(w1)+1)]
    score[:,0] = [0 + i*gap for i in range(len(w2)+1)]
    for i in range(1, len(w2)+1):
        row_l = w2[i-1] # letter at row
        for j in range(1, len(w1)+1):
            col_l = w1[j-1]
            blos_s = blosum62_df.loc[blosum62_df.index == row_l, col_l].values[0]
            diag_s = score[i-1, j-1] + blos_s
            left_s = score[i, j-1] + gap
            top_s = score[i-1, j] + gap
            max_s = max(diag_s, left_s, top_s)
            score[i,j] = max_s
            if max_s == diag_s:
                arrow[i,j] = 'diag'
            elif max_s == left_s:
                arrow[i,j] = 'left'
            else:
                arrow[i,j] = 'top'
            
            if max_s == diag_s == left_s == top_s:
                arrow2[i,j] = 'diag&left&top'
            if max_s == diag_s == left_s:
                arrow2[i,j] = 'diag&left'
            elif max_s == diag_s == top_s:
                arrow2[i,j] = 'diag&top'
            elif max_s == diag_s:
                arrow2[i,j] = 'diag'
            elif max_s == left_s:
                arrow2[i,j] = 'left'
            else:
                arrow2[i,j] = 'top'

    A = ''
    B = ''
    flags = np.array([['' for _ in range(len(w1)+1)] for _ in range(len(w2)+1)], dtype=object)

    while (i!=0) | (j!=0):
        flags[i,j] = arrow[i,j]
        if arrow[i,j] == 'diag':
            a = w1[j-1]
            b = w2[i-1]
            i = i-1
            j = j-1
        elif arrow[i,j] == 'left':
            a = w1[j-1]
            b = '_'
            j = j-1
        else:
            a = '_'
            b = w2[i-1]
            i = i-1
        A = A + a 
        B = B + b

    score = score.astype(int)
    plt.figure(figsize=(25, 25))

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Create a heatmap of the score matrix
    cax = ax.matshow(score, cmap='Blues', alpha=0)

    # Set column labels (w1, starting from the second column)
    ax.set_xticks(np.arange(1, len(w1) + 1))
    ax.set_xticklabels(list(w1))

    # Set row labels (w2, starting from the second row)
    ax.set_yticks(np.arange(1, len(w2) + 1))
    ax.set_yticklabels(list(w2))

    # Overlay numbers, arrows from arrow2 matrix, and bold flags
    for i in range(score.shape[0]):
        for j in range(score.shape[1]):
            # Display the number from the score matrix
            ax.text(j, i, f'{score[i, j]}', va='top', ha='left', color='black', fontsize=7)
            if i == 0 and j == 0:
                continue
            if flags[i,j]:
                if 'left' in arrow2[i, j] and 'diag' in arrow2[i,j]:
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                    ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=12)
                elif 'top' in arrow2[i,j] and 'diag' in arrow2[i, j]:
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                    ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=12)
                elif arrow[i, j] == 'diag':
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                elif arrow2[i,j] == 'left':
                    ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                else:
                    ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
            else:
                if 'left' in arrow2[i, j] and 'diag' in arrow2[i,j]:
                    ax.text(j, i, '↖←', va='bottom', ha='center', color='black', fontsize=12)
                elif 'top' in arrow2[i,j] and 'diag' in arrow2[i, j]:
                    ax.text(j, i, '↖↑', va='bottom', ha='center', color='black', fontsize=12)
                elif arrow2[i, j] == 'left':
                    ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=12)
                elif arrow2[i, j] == 'top':
                    ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=12)
                else:
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=12)

            # Overlay flags in bold, based on the values in flags matrix
            #if flags[i, j] == 'diag':
             #   ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=18, fontweight='bold')
            #elif flags[i, j] == 'left':
             #   ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=18, fontweight='bold')
            #elif flags[i, j] == 'top':
             #   ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=18, fontweight='bold')

    # Set column and row labels (starting from the second position)
    ax.set_xticks(np.arange(len(w1) + 1), minor=False)
    ax.set_yticks(np.arange(len(w2) + 1), minor=False)

    # Ensure the plot displays correctly
    plt.show()
    
    print("Global Alignment")
    print(A[::-1])
    print(B[::-1])
```

```python
# Smith Waterman algorithm

def smith_waterman(w1, w2, gap):
    #if len(w1)<len(w2):
     #   w1, w2 = w2, w1
    
    print("Local Alignment using Smith-Waterman Algorithm")
    print(f"words: {w1}, {w2} and gap: {gap}")

    score = np.zeros((len(w2) + 1, len(w1) + 1))
    arrow = np.array([['' for _ in range(len(w1)+1)] for _ in range(len(w2)+1)], dtype=object)
    arrow[0,0] = ''
    arrow2 = arrow.copy()
    for i in range(1, len(w2)+1):
        row_l = w2[i-1] # letter at row
        for j in range(1, len(w1)+1):
            col_l = w1[j-1]
            blos_s = blosum62_df.loc[blosum62_df.index == row_l, col_l].values[0]
            diag_s = score[i-1, j-1] + blos_s
            left_s = score[i, j-1] + gap
            top_s = score[i-1, j] + gap
            max_s = max(diag_s, left_s, top_s)
            if max_s>0:
                score[i,j] = max_s
                if max_s == diag_s == left_s:
                    arrow[i,j] = 'diag'
                    arrow2[i,j] = 'diag&left'
                elif max_s == diag_s == top_s:
                    arrow[i,j] = 'diag'
                    arrow2[i,j] = 'diag&top'
                elif max_s == diag_s:
                    arrow[i,j] = 'diag'
                    arrow2[i,j] = 'diag'
                elif max_s == left_s:
                    arrow[i,j] = 'left'
                    arrow2[i,j] = 'left'
                else:
                    arrow[i,j] = 'top'
                    arrow2[i,j] = 'top'
            else:
                score[i,j] = 0

    # Find the index of the maximum element
    max_index = np.argmax(score)
    # Convert the flat index to row and column coordinates
    i, j = np.unravel_index(max_index, score.shape)
    A = ''
    B = ''
    flags = np.array([['' for _ in range(len(w1)+1)] for _ in range(len(w2)+1)], dtype=object) 

    while ((i!=0) | (j!=0)) & (score[i,j]!=0):
        flags[i,j] = arrow[i,j]
        if arrow[i,j] == 'diag':
            a = w1[j-1]
            b = w2[i-1]
            i = i-1
            j = j-1
        elif arrow[i,j] == 'left':
            a = w1[j-1]
            b = '_'
            j = j-1
        else:
            a = '_'
            b = w2[i-1]
            i = i-1
        A = A + a 
        B = B + b

    score = score.astype(int)

    plt.figure(figsize = (20, 20))
    # Create a figure and axis
    fig, ax = plt.subplots()

    # Create a heatmap of the score matrix
    cax = ax.matshow(score, cmap='Blues', alpha = 0)

    # Add color bar for the heatmap
    #plt.colorbar(cax)

    # Set column labels (w1, starting from the second column)
    ax.set_xticks(np.arange(1, len(w1) + 1))
    ax.set_xticklabels(list(w1))

    # Set row labels (w2, starting from the second row)
    ax.set_yticks(np.arange(1, len(w2) + 1))
    ax.set_yticklabels(list(w2))

    # Overlay numbers and arrows on the cells
    for i in range(score.shape[0]):
        for j in range(score.shape[1]):
            # Display the number from the score matrix
            ax.text(j, i, f'{score[i, j]}', va='top', ha='left', color='black', fontsize=7)

            # Overlay an arrow based on the arrow matrix
            if flags[i,j] != '':
                if 'left' in arrow2[i, j] and 'diag' in arrow2[i,j]:
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                    ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=12)
                elif 'top' in arrow2[i,j] and 'diag' in arrow2[i, j]:
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                    ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=12)
                elif arrow[i, j] == 'diag':
                    ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                elif arrow2[i,j] == 'left':
                    ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
                else:
                    ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=16, fontweight = 'bold')
            else:
                if score[i,j] > 0:
                    if 'left' in arrow2[i, j] and 'diag' in arrow2[i,j]:
                        ax.text(j, i, '↖←', va='bottom', ha='center', color='black', fontsize=12)
                    elif 'top' in arrow2[i,j] and 'diag' in arrow2[i, j]:
                        ax.text(j, i, '↖↑', va='bottom', ha='center', color='black', fontsize=12)
                    elif arrow[i, j] == 'left':
                        ax.text(j, i, '←', va='bottom', ha='center', color='black', fontsize=12)
                    elif arrow[i, j] == 'top':
                        ax.text(j, i, '↑', va='bottom', ha='center', color='black', fontsize=12)
                    else:
                        ax.text(j, i, '↖', va='bottom', ha='center', color='black', fontsize=12)

    # Set column and row labels (starting from the second position)
    ax.set_xticks(np.arange(len(w1) + 1), minor=False)
    ax.set_yticks(np.arange(len(w2) + 1), minor=False)

    # Ensure the plot displays correctly
    plt.show()

    print("Local Alignment")
    print(A[::-1])
    print(B[::-1])
```
